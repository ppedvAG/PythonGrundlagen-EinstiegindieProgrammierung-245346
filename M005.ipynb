#### N.A.

Keine automatische Lösung -> Boolean Masken

```python
data == "N.A."  # Dadurch das N.A. im Datenset verteilt ist, kann hier nicht einfach ein Rechteck genommen werden
```

```python
(data == "N.A.").value_counts()  # Jede Spalte mit mind. einem True wird hier gelistet
```

```python
data["FertRate"] == "N.A."  # Alle N.A. Zeilen in FertRate finden
```

```python
data[data["FertRate"] == "N.A."]["FertRate"] = 0
```

```python
data[data["FertRate"] == "N.A."].index
```

```python
data.loc[data[data["FertRate"] == "N.A."].index, "FertRate"] = 0
```

```python
data
```

Nachdem noch Zwei weitere Spalten Fehler haben, bietet es sich an, eine Funktion zu schreiben

```python
def fixNA(colName, newValue):
    b = data[colName] == "N.A."  # Fehlerhafte Zeilen finden
    f = data[b]  # Fehlerhafte Zeilen herausfiltern
    i = f.index  # Index der fehlerhaften Zeilen nehmen
    data.loc[i, colName] = newValue  # Fehlerhafte Werte überschreiben
```

```python
fixNA("MedAge", 0)
```

```python
for col in data.columns:
    fixNA(col, 0)
```

```python
data
```

#### Prozentzeichen

Hier wird string-replace verwendet

" %" -> ""

```python
data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data["YearlyChange"] = data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data
```

Funktion

```python
def fixPct(colName):
    data[colName] = data[colName].replace(" %", "", regex=True)
```

```python
for col in data.columns:
    fixPct(col)
```
