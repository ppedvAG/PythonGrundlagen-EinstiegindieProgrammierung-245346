## Matplotlib

Hauptwerkzeug zur Darstellung von Daten

https://matplotlib.org/stable/plot_types/index.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # Immer Unterpaket pyplot importieren
```

```python
data = pd.read_csv("Data/PopulationDataFertig.csv", index_col="Country")
```

### Globale Graphen vs. Objektorientierte Graphen

```python
data["WorldShare"].head(10).plot()  # Über Pandas selbst
```

#### Globaler Modus

```python
# Über plt.<Befehl> wird hier ein Graph gezeichnet
plt.figure(figsize=(12, 12))
plt.pie(data["WorldShare"].head(10), labels=data.index[:10])
plt.legend(loc="upper right")
plt.show()  # Mit plt.show() wird der Graph abgeschlossen

None
```

#### Objektorientierter Modus

```python
fig, ((a1_1, a1_2), (a2_1, a2_2)) = plt.subplots(2, 2)
a1_1.plot(data["Pop"].head(10))
a1_2.plot(data["Pop"][10:20])
a2_1.plot(data["Pop"][20:30])
a2_2.plot(data["Pop"][30:40])
```
